```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
df_cleaned = df.drop(columns=['id','date','zipcode'])
```

```python
#5.1
stats = pd.DataFrame({
    'Average': df_cleaned.mean(),
    'Min': df_cleaned.min(),
    'Max': df_cleaned.max(),
    'Variance': df_cleaned.var()
})
print('Lowest Average: ', stats['Average'].idxmin(), stats['Average'].min())
print('Highest Average: ', stats['Average'].idxmax(), stats['Average'].max())
print('Lowest Variance: ', stats['Variance'].idxmin(), stats['Variance'].min())
print('Highest Variance: ', stats['Variance'].idxmax(), stats['Variance'].max())
```

```python
#5.2
y = df['price']
df_corr = df_cleaned.drop(columns=['price'])
corr = df_corr.corrwith(y)
corr_table = corr.sort_values(ascending=False).to_frame(
    name="correlation_with_price"
)
print(corr_table)
print()
print('All features are positively correlated with the response')
print('Highest Positive Correlation: ', corr_table['correlation_with_price'].idxmax())
```
