```python
import numpy as np
```

```python
np.random.seed(0)
```

```python
#Each of 3 matrices
matrices = []

for i in range(3):
    M = np.random.randint(-10, 11, size=(3,3))
    matrices.append(M)
    print(f"Matrix {i}:\n", M)
```

```python
#The inverses computed with the package
inverses = []

for i, M in enumerate(matrices):
    try:
        M_inv = np.linalg.inv(M)
        inverses.append(M_inv)
        print(f"Inverse Matrix {i}:\n", M_inv)
    except np.linalg.LinAlgError:
        inverses.append(None)
        print(f"Matrix{i} has no inverse.")
```

```python
#Compute the product of each matrix with its inverse to check that you obtain the identity matrix.
for i in range(len(matrices)):
    if inverses[i] is not None:
        product = matrices[i] @ inverses[i]
        identity_check = np.round(product, 10)
        print(f"Product of Matrix {i} and its Inverse:\n", identity_check.astype(int))
    else:
        print(f"Matrix {i} has no inverse, so no verification is possible.")
```
